# Spotify Data Analysis

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Analysing "tracks" dataset

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# Null values in the dataset
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# 10 Least popular songs on spotify
sorted_df = df_tracks.sort_values('popularity', ascending=True).head(10)  
sorted_df
```

```python
# Descriptive Statistics of "tracks" dataset
df_tracks.describe().transpose()
```

```python
# 10 most popular songs with popularity > 90

most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular.head(10)
```

```python
# Setting "release_date" as index of dataset

df_tracks.set_index('release_date',inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Print the name of artist of 19th row of dataset

df_tracks[['artists']].iloc[18]
```

```python
# Convert the "duration_ms" to "duration" in seconds

df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace =True, axis = 1)
df_tracks.head()
```

### Correlation among the Columns of "tracks" dataset
 

```python
df_tracks_temp = df_tracks.drop(['key','mode','explicit'], axis = 1) # Dropping unwanted columns from the dateset
corr_df = df_tracks_temp.corr(method='pearson')

# Plotting the correlation values 
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot=True, fmt = "0.1g",vmin=-1,vmax=1,center=0, cmap="inferno",linecolor='black',linewidths=1)
heatmap.set_title("Correaltion HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation =90)

plt.show()
%matplotlib inline
```

```python
# Creating regression plots with 0.4% of data

sample_df = df_tracks.sample(int(0.004*len(df_tracks))) # Sampling data
print(len(sample_df)) # Length of sampled data
```

```python
# Regression Plot I - Loudness Vs Energy Correlation

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x="energy", y="loudness", color="c").set_title("Loudness Vs Energy Correlation")
```

```python
# Regression Plot II - Popularity Vs Acousticness

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x="acousticness", y="popularity", color="b").set_title("Popularity Vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index
years = df_tracks.dates.dt.year

# Distribution plot for number of songs per year
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
# Bar Plot to analize Average song duration per year 

total_dr = df_tracks.duration
fig, ax = plt.subplots(figsize=(18,7))
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title = "Years Vs Duration")
plt.xticks(rotation= 90)

plt.show()
%matplotlib inline
```

```python
# Line Plot to analize Average song duration each year 

sns.set_style(style='whitegrid')
fig, ax = plt.subplots(figsize=(15,6))
fig= sns.lineplot(x=years, y=total_dr, ax=ax).set(title = "Years Vs Duration")
plt.xticks(rotation= 60)

plt.show()
%matplotlib inline
```



## Analysing "SpotifyFeatures" dataset

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
# Bar Plot of duration of songs in different genres

plt.figure(figsize=(12,8))
sns.barplot(y ='genre', x='duration_ms', data = df_genre).set(title = "Duration of the Songs in Different Genres")
plt.ylabel("Gneres")
plt.xlabel("Duration in milli-seconds")

plt.show()
%matplotlib inline
```

```python
# Top 5 most popular genres 

famous = df_genre.sort_values("popularity", ascending = False).head(10) 
sns.color_palette("crest", as_cmap=True)
plt.figure(figsize=(10,5))
sns.barplot(y='genre', x ='popularity', data=famous).set(title="Top 5 Popular Genres")

plt.show()
%matplotlib inline
```
